## clear content

```python
def filter_contents(contents):
    # 去除html标签
    contents = [re.sub('</?[^>]+>', '', content) for content in contents]

    # 去除style风格
    contents = [re.sub('style=".*"', '', content) for content in contents]
    
    # 替换成句号
    contents = [content.replace(u'\r\n', u'。').replace(u'\n', u'。').replace(u'\t', u'') for content in contents]
    
    # 替换空格
    contents = [content.replace(u'\u3000', u'').replace(u'\xa0', u'').replace(u' ', u'') for content in contents]

    return contents
```

## cut sentences

```python
def cut_paragraph(para):
    para = re.sub('([。！？\?])(?!”)',r"\n",para) # 单字符断句符
    para = re.sub('(\.{6})(?!”)',r"\n",para) # 英文省略号
    para = re.sub('(\…{2})(?!”)',r"\n",para) # 中文省略号
    para = re.sub('(。”)','。”\n',para)   # 把分句符\n放到双引号后，注意前面的几句都小心保留了双引号
    para = re.sub('[^(（【]+(?=[)）】])','\n',para)  # 匹配括号里的内容，如（摄影：亚瑟夫/人民网）
#     para = re.sub('[^【]+(?=[)）])','\n',para) 
    para = para.rstrip()       # 段尾如果有多余的\n就去掉它
    para = para.replace('（', '').replace('）', '').replace('(', '').replace(')', '')
    para = para.replace('【', '').replace('】', '').replace('。”', '')
    sent_list = para.split("\n")
    sent_list = [sent for sent in sent_list if sent != ""]
    return sent_list
```

## 分割句子成片段

```python
def cut_sentence(sentence):
    '''一句话分割成片段（bits_list）'''
    sentence = re.sub('([，|；|：“])(?!”)',r"\n",sentence) # 单字符断句符
    sentence = sentence.replace('”', '').replace('“', '')
    bits_list = sentence.split("\n")
    bits_list = [sent.strip() for sent in bits_list if sent != ""]
    return bits_list
```

```python
bits_weight_list = []
bits_dict  = {}

for [person_weight, textrank_weight,c_id, s_id, sentence] in sentences_weight_list:
    bits_list = cut_sentence(sentence)
    bits_list = filter(lambda s: len(s) > 4, bits_list)
    for b_id, sent in enumerate(bits_list):
        # person_weight, textrank_weight,, c_id, s_id, b_id, sent
        bits_weight_list.append([person_weight, textrank_weight,c_id, s_id, b_id, sent]) 
        if sent not in bits_dict:
            bits_dict[sent] = 1
        else :
            bits_dict[sent] = bits_dict[sent] + 1
        
print('bits_weight_list: ', len(bits_weight_list))
bits_weight_list
```
