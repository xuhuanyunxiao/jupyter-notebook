## feature: 相似性

```python
used_data = title
cor_table = np.zeros(shape=(len(used_data),len(used_data)))

for i, t1 in enumerate(used_data):
    hash1 = simhash(t1)
    for j, t2 in enumerate(used_data):
        hash2 = simhash(t2)
        if i == j :  # 对角线上
            sim = 1
        elif i > j :
            sim = cor_table[j, i]  # 左下角--对应 右上角
        else :
            sim = hash1.similarity(hash2)  # 右上角，计算
#         print (i, j , sim)
        cor_table[i, j] = sim
# cor_table
```

```python
used_data = title
cor_table = np.zeros(shape=(len(used_data),len(used_data)))
print 'used_data: ', len(used_data)
print 'cor_table: ', cor_table.shape

k = 0
for i, t1 in enumerate(used_data):
    hash1 = simhash(t1)
    for j, t2 in enumerate(used_data):
        k += 1
        if k % 50000 ==0: print 'k: ', k
            
        hash2 = simhash(t2)
        if i == j :  # 对角线上
            sim = 1
        elif i > j :
            sim = cor_table[j, i]  # 左下角--对应 右上角
        else :
            sim = hash1.similarity(hash2)  # 右上角，计算
#         print (i, j , sim)
        cor_table[i, j] = sim
# cor_table
```
