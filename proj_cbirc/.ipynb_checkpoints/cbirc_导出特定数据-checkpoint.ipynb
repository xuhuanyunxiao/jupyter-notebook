## keyword

```python
f = open('data/国寿关键词.txt', 'r', encoding = 'utf-8')
keyword_list = []
for line in f.readlines():
    keyword_list.append(line.replace('\n', ""))
f.close()

print(keyword_list)
```

```python
keyword_data = pd.DataFrame()
for index, keyword in enumerate(keyword_list):
    if keyword != '':
        sql_keyword = "SELECT t1.id, t1.keyword  \
                            FROM db_keywords_gather t1 \
                                where locate('{0}', t1.keyword)".format(keyword)
        tmp_data = pd.read_sql(sql_keyword, engine)

        if tmp_data.shape[0] > 0:
            print('------- ', index)
            print('keyword: ', keyword)
            print(tmp_data.shape)
            print(tmp_data)

            keyword_data = pd.concat([keyword_data, tmp_data], axis = 0)
    
keyword_data = keyword_data.drop_duplicates('id')
print(keyword_data.shape)
keyword_data.head()
```

```python
keyword_data['keyword_c'] = keyword_data['keyword'].apply(lambda x: x.replace('/', '_'))
keyword_data.head()
```

```python
keyword_c_list = keyword_data['keyword_c'].tolist()
print(len(keyword_c_list))
print(keyword_c_list)
```

```python
sel_str = '0、1、4、6、7、10、11、12、18、65'
sel_list = [int(s) for s in sel_str.split('、')]
sel_keyword_list = [keyword_c_list[s] for s in sel_list]
sel_keyword_list_time = ['2018-07-04','2018-07-03','2018-07-01',
                         '2018-06-26','2018-06-25','2018-06-01',
                        '2018-06-27','2018-07-03','2018-06-01','2018-07-09']
sel_keyword_list
```

## 查找数据

```python
human_data = []
for index, keyword in enumerate(keyword_c_list):
    print('----- index: %s, keyword: %s'%(index, keyword))
    
    sql_human_count = "SELECT count(t2.id) \
                        FROM db_docinfo_trade t1, db_docinfo t2\
                            where t1.keywords = '{0}' \
                                    and date_format(t2.publishtime, '%%Y-%%m-%%d') >= '2018-07-01'\
                                    and t1.urlhash = t2.urlhash".format(keyword)
    tmp_data_count = pd.read_sql(sql_human_count, engine)
    
    if tmp_data_count.iloc[0,0] > 1000:
        day_str = '2018-07-01'
    else :
        day_str = '2018-06-01'
    
    sql_human = "SELECT t2.id, t2.site_name, t2.author, t2.publishtime, t2.group_id, \
                        t2.title, t3.text as content\
                        FROM db_docinfo_trade t1, db_docinfo t2, db_docinfo_text t3 \
                            where t1.keywords = '{0}' \
                                    and date_format(t2.publishtime, '%%Y-%%m-%%d') >= '{1}'\
                                    and t1.urlhash = t2.urlhash \
                                    and t1.urlhash = t3.urlhash \
                            order by t2.publishtime desc \
                            limit 1000".format(keyword, day_str)
    tmp_data = pd.read_sql(sql_human, engine)
    
    if len(keyword) > 71:keyword = keyword[:70]
    tmp_data.to_excel('result/%s(%s)-%s-%s.xlsx'%(index, tmp_data.shape[0], day_str, keyword), index = False)
    human_data.append([index, keyword, day_str, tmp_data.shape[0]])
    
human_data = pd.DataFrame(human_data, columns = ['index', 'keyword', 'day_start', '数据量'])
human_data.to_excel('result/导出数据统计-人寿-广发行-20180710.xlsx', index = False)
```

```python
human_data = []
for index, keyword in enumerate(sel_keyword_list):
    print('----- index: %s, keyword: %s'%(index, keyword))
    
    day_str = sel_keyword_list_time[index]
    
    sql_human = "SELECT t2.id, t2.site_name, t2.author, t2.publishtime, t2.group_id, \
                        t2.title, t3.text as content\
                        FROM db_docinfo_trade t1, db_docinfo t2, db_docinfo_text t3 \
                            where t1.keywords = '{0}' \
                                    and (date_format(t2.publishtime, '%%Y-%%m-%%d') < '{1}')\
                                    and t1.urlhash = t2.urlhash \
                                    and t1.urlhash = t3.urlhash \
                            order by t2.publishtime desc \
                            limit 15000".format(keyword, day_str)
    tmp_data = pd.read_sql(sql_human, engine)
    
    if len(keyword) > 71:keyword = keyword[:70]
    tmp_data.to_excel('result/sel_%s(%s)-%s-%s.xlsx'%(index, tmp_data.shape[0], day_str, keyword), index = False)
    human_data.append([index, keyword, day_str, tmp_data.shape[0]])
    
human_data = pd.DataFrame(human_data, columns = ['index', 'keyword', 'day_start', '数据量'])
human_data.to_excel('result/sel_导出数据统计-人寿-广发行-20180710.xlsx', index = False)
```
