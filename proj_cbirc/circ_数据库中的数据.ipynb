# 过模型后相关数据表

## wise_web_docinfo （业务表）
> 相关数据表标题

+ 字段：
- - gather_type
> - 0 ---- 系统全面采集
> - 1 ---- 人工补录（未过算法，人工判断为相关）
> - 2 ---- 人工校正（过了算法，人工修改内容）
> - 3 ---- 批量导入（Excel导入，未过算法，人工判断为相关）
- - noise_flag 
> - -- 1（有用） 是人工判断为噪音的数据，即算法判断为相关而业务人员判断为不相关
> - -- 0 默认是算法判断为相关数据，但无法分别业务人员也做如此判断
- - publistime（发布时间）
- - gathertime（采集时间）    
- - inserttime（入库时间，入 业务表 时间）

```python
sql_c = 'select * from wise_web_docinfo limit 5'
data = get_sql_data(sql_c)
desc_data(data)
```

```python
sql_c = 'select gather_type, count(gather_type) as gather_type_num \
            from wise_web_docinfo \
                group by gather_type'
data = get_sql_data(sql_c)
pd.DataFrame(data)
```

```python
sql_c = 'select noise_flag, count(noise_flag) as noise_flag_num \
            from wise_web_docinfo \
                group by noise_flag'
data = get_sql_data(sql_c)
pd.DataFrame(data)
```

```python
# 自动输出 HTML 文件
def output_HTML(read_file, output_file):
    from nbconvert import HTMLExporter
    import codecs
    import nbformat
    exporter = HTMLExporter()
    # read_file is '.ipynb', output_file is '.html'
    output_notebook = nbformat.read(read_file, as_version=4)
    output, resources = exporter.from_notebook_node(output_notebook)
    codecs.open(output_file, 'w', encoding='utf-8').write(output)
              
current_file = os.getcwd() + '\\circ_数据库中的数据.ipynb'
html_file_folder = r'D:\XH\OneDrive\网智天元\4 银保监会\分析结果'
output_file = html_file_folder + '\\circ_数据库中的数据.html' 
output_HTML(current_file, output_file)
```
